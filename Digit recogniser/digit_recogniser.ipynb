```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import tensorflow as tf
import os
```

```python
img = cv2.imread("train/0/img_4.jpg", cv2.IMREAD_COLOR)[:,:,0]
```

```python
print(os.listdir("train/0"))
```

```python
#loading the training dataset
X=[]
Y=[]
for i in os.listdir("train"):
    c=0
    for j in os.listdir(f"train/{i}"):
        img=cv2.imread(f"train/{i}/{j}")[:,:,0]
        flattern_list=[k for sub in img for k in sub]
        Y.append(i)
        X.append(flattern_list)
        if c==100:
            break
        c+=1
X=np.array(X)
Y=np.array(Y)
```

```python
#building the model
model=tf.keras.Sequential()
model.add(tf.keras.layers.Dense(128,activation='relu'))
model.add(tf.keras.layers.Dense(128,activation='relu'))
model.add(tf.keras.layers.Dense(128,activation='relu'))
model.add(tf.keras.layers.Dense(10,activation='softmax'))

model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X,Y,epochs=5)
model.save('digit_recogniser')
```
